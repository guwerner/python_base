```python
df = pd.read_csv("./data/data.csv")
print("First 5 Rows:\n", df.head(5))
print("Last 5 Rows:\n", df.tail(5))
print("Shape:", df.shape)
print("Column Names:", df.columns)
print("Data Types:\n", df.dtypes)
print("Statistical Summary:\n", df.describe())
print("Number of Unique Columns:\n", df.nunique())
print("Number of Missing Values per Column:\n", df.isnull().sum()) #sum = total number
print("Memory Usage:", df.memory_usage(deep=True))

#Data Structure Management

df.rename(columns={'Duration':'Time Span'}, inplace=True)
df = df.iloc[:, [1, 0, 2, 3, 4]]
df['Calories2'] = df['Calories']
del df['Calories2']
df_copy = df.copy()
df = df.reset_index(drop=True)
#df = df.set_index('Time Span')
#df = df.sort_index(ascending=False)

#Missing Data Handling

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].fillna('Unknown')
df = df.fillna(0)
print(df)
```
